softmax不改变分类预测结果，只是生成分类预测的概率

交叉熵损失函数


$y^{(i)}$ = [一个0/1向量，第i个样本假如属于第k类，则第k个位置为1，其他位置为0]

$\hat{y}^{{i}}$= [一个连续值向量，各元素和等于1.0，是模型预测值经过softmaxt处理后的结果，即的属于各个类别的概率]  
$\hat{y}^{{i}}_{j}$表示模型预测样本i属于第j类的概率

总结：
* 最小化交叉熵损失函数等价于最大化训练数据集所有标签类别的联合预测概率
* 交叉熵适合衡量两个概率分布的差异
